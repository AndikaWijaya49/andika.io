```python
df = pd.read_csv ('10000 Sales Records.csv')
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
X = df [['Item Type', 'Sales Channel', 'Order Priority', 'Country']]
Y = df ['Region']
```

```python
le = LabelEncoder()
x_encoded = X.apply(le.fit_transform)
y_encoded = le.fit_transform(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split (x_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
accuracy_score (y_test, y_pred)
classification_report(y_test, y_pred)
```

```python
plt.figure (figsize=(15, 8))
plot_tree(model, feature_names= X.columns, class_names= le.classes_, filled= True)
plt.title("Decision Tree untuk Prediksi Region")
```
